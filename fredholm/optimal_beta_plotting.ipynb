```python
import os

from IPython.display import display

from fredholm_plotting import *
from fredholm_utils import estimated_b_function_matrix

def no_gibbs_make_plots(filepath, range_linspace=[-5,5], plotting_y_lim=None, plotting_x_lim=None, data_histogram=False, optimal_beta='y', dirpath='fredholm_results/results20' , **kwargs):
    # find all the files in the directory given

    dict_list = get_results_dictionaries([filepath], dirpath=dirpath)
    meta = metadata_table(dict_list)
    display(meta)

    if data_histogram:
        plot_data_histogram(dict_list, **kwargs)

    b_mat = estimated_b_function_matrix(
        range_linspace=range_linspace,
        init_data=dict_list[0]['init_data'],
        kernel_name = dict_list[0]['meta_data']['kernel_name'],
        known_b=dict_list[0]['known_b'],
        y_domain=[dict_list[0]['meta_data']['y_domain'][0], dict_list[0]['meta_data']['y_domain'][1]],
        int_n = 5000
    )
    plot_bbar_estimates(dict_list, b_mat, y_lim=plotting_y_lim, x_lim=plotting_x_lim, optimal_beta=optimal_beta)
```

```python
# What do we want to do here?
# Essentially, I want to generate data for a longer run (say t_end=500, instead of 100) 
# then calculate the optimal values of beta (using the b_mat)
# no gibbs required.

no_gibbs_make_plots('b2_T50_diff1p5.pkl', data_histogram=True, bins=30, optimal_beta='only', dirpath='fredholm_results/results20/')
```
