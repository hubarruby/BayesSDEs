```python
#it's the same, nice
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define the parameters
loc = -1
scale = 2
a = 0.5
b = 0.5

# Generate x values
x = np.linspace(beta.ppf(0.01, a, b, loc=loc, scale=scale),
                beta.ppf(0.99, a, b, loc=loc, scale=scale), 100)
pdf = beta.pdf(x, a, b, loc=loc, scale=scale)
plt.plot(x, pdf, label='Beta PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of Beta Distribution')
plt.legend()
plt.show()
```
